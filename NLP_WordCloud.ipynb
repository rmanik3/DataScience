```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import docx2txt

%matplotlib inline
```

```python
# result = docx2txt.process("C:/Users/karthik/Desktop/Ramkarthik_Resume_2019.docx")
from docx2python import docx2python
doc_output = docx2python('/Users/Karthik/Desktop/Canada PR/Ramkarthik_Resume_2020_1.docx')
```

```python
stopwords = set(STOPWORDS)
stopwords.update(["Ramkarthik","ramkarthikirtt@gmail.com"])
print(stopwords)
```

```python
# Create and generate a word cloud image:
wordcloud = WordCloud(stopwords=stopwords, background_color="white",max_font_size=75,max_words=25).generate(doc_output.text)
```

```python
# Display the generated image:
fig, ax = plt.subplots(figsize=(15,10 ))
plt.imshow(wordcloud, interpolation='bilinear')#what is bilinear interpolation ?
plt.axis("off")
plt.show()
```

```python
#Creating custom word cloud images based on image templates
# Read the template image
from PIL import Image
# Image.open(r"C:/Users/karthik/Desktop/KT & Learning/ML/NLP/Image_C.jpg")
mask = np.array(Image.open(r"/Users/Karthik/Desktop/KT & Learning/ML/NLP/Image_C.jpg"))
```

```python
wordcloud_img = WordCloud(stopwords=stopwords, background_color="white", mode="RGBA", max_font_size=75, max_words=25, mask=mask).generate(doc_output.text)

plt.figure(figsize=[10,10])
plt.imshow(wordcloud_img, interpolation="bilinear") 
plt.axis("off")
```
