```python
#Creating the fitness score function by taking the max value of the matches in 92 possible solutions
def fitness_score(each_pop_comb):
    final_fit_score = []
    for tgt_list in range(len(non_fight_92)):
        fit_score = 0
        for each_digit in range(8):
            if each_pop_comb[each_digit] == non_fight_92[tgt_list][each_digit]:
                fit_score = fit_score + 1
        final_fit_score.append(fit_score)
    return max(final_fit_score)
```

```python
##Define the configuration for population and fitness score
from numpy.random import choice
import random
import numpy as np
import pandas as pds
secure_random = random.SystemRandom()
populationSize = 200
populationData = []
fitnessScore = []#fitness score is based on the strength of match from the 92 combinations
secure_random = random.SystemRandom()
chess_board_index = [0, 1, 2, 3, 4, 5, 6, 7]
```

```python
#Generating the population 
for each in range(populationSize):
  randomData = []
  fitScore = 0
  for each_digit in range(len(chess_board_index)):
    selectedData = secure_random.choice(chess_board_index)
    randomData.append(selectedData)
  populationData.append(randomData)
  fitnessScore.append(fitness_score(randomData))
population_fitness = pds.concat([pds.Series(populationData), pds.Series(fitnessScore)], axis=1)
print(population_fitness.head(10))
```

```python
prob_density = []
for each_pop in range(len(populationData)):
  prob_density.append(fitnessScore[each_pop]/len(chess_board_index))
prob_df = pds.DataFrame({'String':populationData,'FitnessScore':fitnessScore,'Probability':prob_density})
prob_df = prob_df.sort_values(['Probability'],ascending=False)
prob_df = prob_df.reset_index(drop=True)
print(prob_df.head())
```

```python
#setting up Cross Over and Mutation Configuration
#crossover point is usually the middle of the target string - in this case
cross_over_setup = int(0.5*len(chess_board_index))
mutation = 1/len(chess_board_index)
print(cross_over_setup, mutation)
```

```python
#Creating childs from crossing over high probable generations from population
child_generation_size = 1000
for each_child in range(child_generation_size):
    top2 = []
    top2.append(prob_df[0:1]["String"].values[0])#selecting only the sequence
    top2.append(prob_df[1:2]["String"].values[0])
    if fitness_score(top2[0])==len(chess_board_index) | fitness_score(top2[1])==len(chess_board_index):
        print(top2[0],' ',top2[1])
        break#first occurance of Non-Fighting Sequence
#applying cross over
    child1 = top2[0][0:cross_over_setup]+top2[1][cross_over_setup:]
    child2 = top2[1][0:cross_over_setup]+top2[0][cross_over_setup:]
#applying mutation for any one of the digits in the child sequences(1/8)
    child1[random.randint(0,7)] = secure_random.choice(chess_board_index)
    child2[random.randint(0,7)] = secure_random.choice(chess_board_index)
#adding childs to the original population
    populationData.append(child1)
    populationData.append(child2)
    recalc_fitness = []
    newPopulationSize = len(populationData)
    for each in range(newPopulationSize):
        fitScore = fitness_score(populationData[each])
        recalc_fitness.append(fitScore)
#calculate the mean fitness Score after the resizing of the population
    prob_density = []
    for each in range(newPopulationSize):
        prob_density.append(recalc_fitness[each]/sum(recalc_fitness))
    prob_df = pds.DataFrame({'String':populationData,'FitnessScore':recalc_fitness,'Probability':prob_density})
    prob_df = prob_df.sort_values(['Probability'],ascending=False)
    prob_df = prob_df.reset_index(drop=True)
    print('Child Generated ',each_child,' ',' Mean Fitness Score ',prob_df["FitnessScore"].mean(),' ', child1,' ',fitness_score(child1),' ', child2, ' ', fitness_score(child2) )
```
